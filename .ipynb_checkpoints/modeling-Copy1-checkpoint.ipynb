```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import (accuracy_score,
                             recall_score,
                             precision_score,
                             f1_score,
                             confusion_matrix,
                             classification_report,
                             plot_confusion_matrix,
                             precision_recall_curve,
                             plot_roc_curve,
                             roc_auc_score)
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, ExtraTreesClassifier

from sklearn.model_selection import GridSearchCV

import warnings
warnings.filterwarnings('ignore')
```

# Data Import

```python
training_df = pd.read_csv('data/training_set_data.csv')
unlabeled_df = pd.read_csv('data/unlabeled_data.csv')
```

### Baseline
Baseline Accuracy of our model should be 0.922479, with a Recall Score (primary) and Precision Score (secondary) as high as possible

```python
# Establish the 0/1 distrobution. This 'null hypothese' percentage (0.922479)
# is the baseline out model should out perform
training_df['uninsured'].value_counts(normalize=True)
```

### Data Cleaning & Feature Engineering

```python
likely_uninsured = training_df[(training_df['marital_status'] == 'never_married') & (training_df['citizen_status'] == 'noncitizen') & 
            (training_df['nativity_status'] == 'foreign_born')]['person_id']
```

```python
training_df['likely_uninsured'] = training_df['person_id'].map(lambda x: 1 if x in likely_uninsured else 0)
```

```python
#Standardize the 'sex' column
def male_female(x):
    x = x.upper()
    if x == 'M':
        return 'MALE'
    elif x == 'F':
        return 'FEMALE'
    else:
        return x

training_df['sex'] = training_df['sex'].map(male_female)
```

```python
to_dummy = ['sex', 'citizen_status', 'nativity_status', 'marital_status', 'school_status', 
            'when_last_worked', 'worked_last_week', 'language']
```

```python
training_df = pd.get_dummies(training_df, columns=to_dummy, drop_first=True)
training_df.head()
```

```python
#Non-numeric, extraneous and target columns
drop_columns = ['uninsured', 'person_id', 'household_id']

# Assign x and y data
X = training_df[[column for column in training_df.columns if column not in drop_columns]]
y = training_df['uninsured']

# Train, Test, Split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)

# Scale the data
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
```

### KNeighborsClassifier

```python
# Instantiate the KNN estimator
knn = KNeighborsClassifier()

# Fit the estimator
knn.fit(X_train_sc, y_train)
```

```python
# cross_val_score(knn, X_train_sc, y_train, cv = 10)
```

```python
# Label predictions
knn_predictions = knn.predict(X_test_sc)

plot_confusion_matrix(knn, X_test_sc, y_test, display_labels=['insured', 'uninsured'])

knn_train_accuracy = knn.score(X_train_sc, y_train)
knn_test_accuracy = knn.score(X_test_sc, y_test)
knn_recall_score = recall_score(y_test, knn_predictions)
knn_precision_score = precision_score(y_test, knn_predictions)
knn_f1_score = f1_score(y_test, knn_predictions)

print(f'LogReg Train Accuracy: {knn_train_accuracy}')
print(f'LogReg Test Accuracy: {knn_test_accuracy}')
print(f'LogReg Test Recall: {knn_recall_score}')
print(f'LogReg Test Precision: {knn_precision_score}')
print(f'LogReg Test F1: {knn_f1_score}')
```

```python
print('ROC AUC Score: {}'.format(roc_auc_score(y_test, knn.predict_proba(X_test_sc)[:,1])))
# ROC curve
plot_roc_curve(knn, X_test_sc, y_test)
# add worst case scenario line
plt.plot([0, 1], [0, 1], color='r', label='Baseline')
# add a legend
plt.legend();
```

### LogisticRegression

```python
#Instantiate our model.
logreg = LogisticRegression()

# Fit the model
logreg.fit(X_train_sc, y_train)

# print(f'Logistic Regression Intercept: {logreg.intercept_}')
# print('Logistic Regression Coefficients:')
# pd.DataFrame(data=tuple(zip(X_train.columns, logreg.coef_[0])), columns = ['Feature', 'Coefficient Value'])
```

```python
# cross_val_score(logreg, X_train_sc, y_train, cv = 10)
```

```python
# Label predictions
logreg_predictions = logreg.predict(X_test_sc)

plot_confusion_matrix(logreg, X_test_sc, y_test, display_labels=['insured', 'uninsured'])

logreg_train_accuracy = logreg.score(X_train_sc, y_train)
logreg_test_accuracy = logreg.score(X_test_sc, y_test)
logreg_recall_score = recall_score(y_test, logreg_predictions)
logreg_precision_score = precision_score(y_test, logreg_predictions)
logreg_f1_score = f1_score(y_test, logreg_predictions)

print('LogReg Train Accuracy: {}'.format(logreg_train_accuracy))
print('LogReg Test Accuracy: {}'.format(logreg_test_accuracy))
print('LogReg Test Recall: {}'.format(logreg_recall_score))
print('LogReg Test Precision: {}'.format(logreg_precision_score))
print('LogReg Test F1: {}'.format(logreg_f1_score))
```

```python
#Show difference in performance based on manually adjusted thresholds
logreg_probs = logreg.predict_proba(X_test_sc)
thresh20_preds = (logreg_probs[:,1] > .2).astype(int)
thresh10_preds = (logreg_probs[:,1] > .1).astype(int)

pd.DataFrame(columns=['0.5 Threshold', '0.2 Threshold', '0.1 Threshold'], index=['accuracy_score', 'recall_score', 'precision_score', 'f1_score'], 
             data=[[accuracy_score(y_test, logreg_predictions), accuracy_score(y_test, thresh20_preds), accuracy_score(y_test, thresh10_preds)],
                  [recall_score(y_test, logreg_predictions), recall_score(y_test, thresh20_preds), recall_score(y_test, thresh10_preds)],
                  [precision_score(y_test, logreg_predictions), precision_score(y_test, thresh20_preds), precision_score(y_test, thresh10_preds)],
                  [f1_score(y_test, logreg_predictions), f1_score(y_test, thresh20_preds), f1_score(y_test, thresh10_preds)]])
```

```python
print('ROC AUC Score: {}'.format(roc_auc_score(y_test, logreg.predict_proba(X_test_sc)[:,1])))
# ROC curve
plot_roc_curve(logreg, X_test_sc, y_test)
# add worst case scenario line
plt.plot([0, 1], [0, 1], color='r', label='Baseline')
# add a legend
plt.legend();
```

```python
plt.figure(figsize=(20, 4))
plt.title('LogReg Probabilities of Being Uninsured')
sns.boxplot(list(logreg_probs[:,1]));


pred_df = pd.DataFrame({'true_values': y_test,
                        'pred_probs': logreg.predict_proba(X_test_sc)[:,1]})

plt.figure(figsize = (20, 5))

# plot distributions of predicted probabilities by actual values
for group in pred_df.groupby('true_values'):
    sns.distplot(group[1], kde = False, bins = 20, label = f'Actual Outcome = {group[0]}')

# Add cutoff line
plt.axvline(.2, color='r')

# Add annotations for TN, FN, TP, FP.
plt.annotate(xy = (0.06, 800), s = 'TN', size = 15)
plt.annotate(xy = (0.1, 200), s = 'FN', size = 15)
plt.annotate(xy = (0.97, 600), s = 'TP', size = 15)
plt.annotate(xy = (0.21, 120), s = 'FP', size = 15)


plt.xlabel('Predicted Probability that Outcome = 1')
plt.ylim((0, 1500))
plt.legend();
```
